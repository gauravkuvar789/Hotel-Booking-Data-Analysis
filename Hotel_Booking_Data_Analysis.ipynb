```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True, errors='coerce')

```

#Explanation :- 
dayfirst=True → tells pandas that the first number is the day, not the month.
errors='coerce' → converts any invalid dates into NaT instead of throwing an error.

```python
df.isnull().sum()
```

First of all remove the company colunm and agent colunm because 1 lakh plus missing value not handle data that's why remove the colunms

```python
df.drop(['company','agent'],axis=1, inplace = True)
df.dropna(inplace=True)  #rows of missing values can use dropna 
```

```python
df = df[df['adr']<5000]
```

Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)  # value_count function return the like category name how many time it is present in colunms

plt.Figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Fix legend (no need to unpack legend_labels if not used)
ax1.legend(title='Is Canceled', bbox_to_anchor=(1, 1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.tight_layout()   # ensures proper spacing
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel ', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label ='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label ='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Extract month from date column
df['month'] = df['reservation_status_date'].dt.month

# Create the figure
plt.figure(figsize=(16, 8))

# Plot the countplot
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Adjust legend and labels
ax1.legend(title='Cancellation Status', labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)

sns.barplot(
    x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(),
    palette='bright'
)

plt.legend(['ADR'], fontsize=20)
plt.show()
```

```python
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.Figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct= '%.2f',labels=top_10_country.index)
plt.show()
```

```python
# For canceled bookings
cancelled_data = df[df['is_canceled'] == 1]
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# For not canceled bookings
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plotting
plt.figure(figsize=(20, 6))
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled')

plt.title('Average Daily Rate', fontsize=20)
plt.xlabel('Reservation Date', fontsize=15)
plt.ylabel('ADR', fontsize=15)
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')& (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20, 6))  
plt.title('Average Daily Rate', fontsize=20)

plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Canceled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled')

plt.xlabel('Reservation Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend(fontsize=20)
plt.tight_layout()  # adjusts spacing nicely
plt.show()
```
